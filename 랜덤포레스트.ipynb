```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

style.use('seaborn-talk')

krfont = {'family':'Malgun Gothic','weight':'bold','size':10}
matplotlib.rc('font',**krfont)
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
huji = pd.read_csv("True.csv",encoding='euc=kr')
huji
```

```python
huji_df = huji.set_index('일자')
```

```python
x = huji[['시점', '거래량.톤.', '거래금액.백만원.', '마늘.면적..ha.',
       'X10a당.생산량..kg.', '생산량..톤.', 'X10a당.생산량.톤', '평균', '평균기온..C.',
       '일강수량.mm.', '평균.풍속.m.s.', '합계.일조시간.hr.', '평균.지면온도..C.',
       '평균.5cm.지중온도..C.']].values
y = huji['도매가격.원.kg.']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

```python

forest = RandomForestRegressor(n_estimators=1000,criterion='mse',random_state=1,n_jobs=-1)
forest.fit(x_train,y_train)
y_train_pred = forest.predict(x_train)
y_test_pred = forest.predict(x_test)

r2_train = r2_score(y_train,y_train_pred)
r2_test = r2_score(y_test,y_test_pred)

print('훈련 R^2: %.3f, 테스트 R^2: %.3f'%(r2_train,r2_test))

forest.feature_importances_
```
